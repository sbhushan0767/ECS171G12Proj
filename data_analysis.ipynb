```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in processed data
df = pd.read_csv('./datasets/credit.csv')
```

```python
# Preview data
df
```

```python
for col in df.columns:
    if (col != 'Credit Score'):
        df.plot.scatter(x=col, y='Credit Score', c='DarkBlue',figsize=(15,8))
```

No correlation indicated:  
- Loan ID  
- Customer ID  
- Loan Status  
- Current Loan Amount****  
- Term  
- Years in Current Job  
- Home Ownership  
- Have had Credit Problems  
- Maximum Open Credit  
- Have had Bankruptcy Before  
- Have had Tax Liens  
- Delinquent Time  

Some correlation indicated:  
- Annual Income: Very mild correlation can be seen with higher annual income corresponding to higher credit scores  
- Purpose: The highest values of purpose seem to correlate mildly to higher credit scores  
- Monthly Debt: Shows noticeable correlation with higher monthly debt corresponding to higher credit scores  
- Years of Credit History: Shows good correlation with more years of credit history corresponding to higher credit scores  
- Number of Open Accounts: Shows good correlation with a higher number of open accounts corresponding to higher credit scores  
- Current Credit Balance: Shows very mild correlation with higher current credit balane correspdoning to higher credit scores  

```python
for col in df.columns:
    if (col != 'Credit Score Range'):
        plt.figure(figsize=(15,8))
        ax = sns.boxplot(x="Credit Score Range", y=col, data=df)
        plt.show()
```

No correlation indicated:  
- Loan ID  
- Customer ID  
- Current Loan Amount  
- Purpose  
- Home Ownership  
- Years in Current Job  
- Have had Credit Problems  
- Have had Bankruptcy Before  
- Have had Tax Liens

Some correlation indicated:  
- Loan Status: Those who were in the 701.2+ usually had the loan status "fully paid"  
- Term: Those below 651.4 had majority of their term values at 0, above 717.8 the majority of term values are 1  
- Years of Credit History: Years of credit history showed mild correlation to those at the highest range of credit scores  
- Number of Open Accounts: The lowest credit score range had the largest 3rd quartile of open accounts   
- Delinquint Time: Shows mild correlation to highest range of credit scores  

Indeterminate:  
- Annual Income: Annual Income shows a reliable correlation to credit score, but there seems to be an issue of outliers here which could be inhibiting proper analysis in this representation  
- Monthly Debt: Outliers are concealing info here as well, but monthly debt shows a mild inverse relationship to credit score
- Current Credit Balance: Shows potential correlation, but outliers are concealing info  
- Maximum Open Credit: Shows potential correlation, but outliers are concealing info  

```python
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(df.corr(), annot=True, linewidths=.5, ax=ax)
```

The heat map indicates the strongest correlations to credit score to be term, current loan amount, and loan status with values of 0.47, 0.11, and 0.15 respectively.

```python
for col in df.columns:
    if (col != 'Credit Score Range'):
        rel = sns.relplot(x="Credit Score Range", y=col, data=df)
        rel.fig.set_size_inches(15,8)
        plt.show()
```

```python
No correlation indicated:  
- Loan ID  
- Customer ID  
- Loan Status  
- Current Loan Amount
- Term  
- Years in Current Job  
- Home Ownership  
- Have had Credit Problems  
- Have had Bankruptcy Before  
- Have had Tax Liens  
- Delinquent Time

Some correlation indicated:  
- Annual Income: Shows good correlation with the highest incomes corresponding to the highest credit scores
- Purpose: We see a mild drop off at the lowest 3 credit score ranges  
- Monthly Debt: Shows a mild correlation with the highest monthly debts corresponding to the highest credit scores (this could be due to increased income allowing increased spending)  
- Years of Credit History: Shows a good correlation with the most years of credit history corresponding to the highest credit scores  
- Number of Open Accounts: Shows a mild correlation with the highest number of accounts corresponding to the highest credit scores   
- Current Credit Balance: Shows a mild correlation with the highest current credit balance corresponding to the highest credit scores   
- Maximum Open Credit: Shows a mild correlation with the highest maximum open credit corresponding to the highest credit scores   
```

```python
Summary:  

The data does not show any major correlations between attributes and credit score, but we were able to discover useful information.  
The scatter and relational plots largely agreed on the attributes which correlated to credit score. The bar plot and heat map  
revealed Term and Loan Status to also be potential useful correlations.  

From all the representations we find the most likely useful attributes to be:
- Annual Income  
- Purpose  
- Years of Credit History    
- Number of Open Accounts (although the bar plot had a confilcting prediction with the scatter and relational plot)  
- Monthly Debt  
- Term  
- Loan Status  

Intuitively, Annual Income and Years of Credit History specifically seem like highly useful attributes for predicting credit score.  
```
